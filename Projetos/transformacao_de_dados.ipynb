# Técnicas de transformações aplicadas a Dados

### Importando pacotes 

```python
# Imports
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings("ignore")
```

### Importando os dados 

```python
# Carrega os dados
dados = pd.read_csv('D:/Projetos_Analise_de_dados/Transformacao_de_variaveis/Dados/dataset.csv')
```

### Aplicando Label Encoding

```python
# Cópia do dataframe original
df_car_le = dados
df_car_le
```

```python
# Cria o objeto encoder
le = LabelEncoder()
```

```python
# Aplica o encoder
df_car_le['Modelo'] = le.fit_transform(df_car_le['Modelo'])
df_car_le
```

### Aplicando One-Hot-Encoding

```python
# Obtém as variáveis dummy
df_dummies = pd.get_dummies(dados['Modelo'], prefix='Modelo')
df_dummies
```

```python
# Concatenamos os dataframes
df_car_ohe = pd.concat([dados, df_dummies], axis = 'columns')
df_car_ohe
```
